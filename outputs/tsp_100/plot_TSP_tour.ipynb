```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
def plot_tsp(xy, tour, ax1, depot=True, arrow_color='green'):
    """
    Plot the TSP tour on matplotlib axis ax1.
    """
    min_loca = np.min(xy, axis=0)
    max_loca = np.max(xy, axis=0)
    
    ax1.set_xlim(min_loca[0]-0.0001, max_loca[0]+0.0001)
    ax1.set_ylim(min_loca[1]-0.0001, max_loca[1]+0.0001)
    
    xs, ys = xy[tour].transpose()
    dx = np.roll(xs, -1) - xs
    dy = np.roll(ys, -1) - ys
    
    # Scatter nodes
    ax1.scatter(xs, ys, s=20, color='blue') 
    ax1.scatter(xs[0], ys[0], s=30, color='red') 
    # Arcs
    if depot:
        # Starting node
        #ax1.scatter([xs[0]], [ys[0]], s=100, color='red')
        qv = ax1.quiver(
            xs, ys, dx, dy,
            scale_units='xy',
            angles='xy',
            scale=1,
            color=arrow_color, 
            width=0.004
        )
        #ax1.set_title('{} nodes with depot, total travel time {:.2f}'.format(len(tour), travel_time))
        ax1.set_title('Tour with depot')
    else:
        qv = ax1.quiver(
            xs[:-1], ys[:-1], dx[:-1], dy[:-1],
            scale_units='xy',
            angles='xy',
            scale=1,
            color='green',
            width=0.003
        )
        #ax1.set_title('{} nodes without depot, total travel time {:.2f}'.format(len(tour), travel_time))
        ax1.set_title('Tour without depot')
    return
```

# REAL - LIFE

```python
with open('../../data/tsp100/test_location.pkl', 'rb') as f:
    location = pickle.load(f)
idx = 1
loca_set  = location[idx][1:]
tour      = range(99)
c = 'black'
fig, axs = plt.subplots(figsize=(10, 10))
#ax1, ax2 = axs[0], axs[1]
plot_tsp(loca_set, tour, axs, arrow_color=c, depot=False)
plt.show()
```
